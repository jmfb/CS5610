## Part 4 - County Analysis

For this we will be using the following libraries...

```python
import pandas as pd
```

Let us load the county business patterns for 2020.

* Source: https://www.census.gov/data/datasets/2020/econ/cbp/2020-cbp.html
* Layout: https://www2.census.gov/programs-surveys/cbp/technical-documentation/records-layouts/2020_record_layouts/county-layout-2020.txt

```python
cbp_df = pd.read_csv("./cbp20co.txt", keep_default_na=True, na_values=["N"]).fillna(0)
cbp_df.head()
```

```python
cbp_df.info()
```

```python
cbp_df.describe()
```

This data is broken down by county and industry.  We will use a rollup by county and just take the sum of all numeric columns (and ignore all categorical ones).

```python
gdf = cbp_df.drop(labels=["fipstate", "naics", "emp_nf", "qp1_nf", "ap_nf"], axis=1).groupby("fipscty").agg("sum")
gdf.head()
```

In order to use this dataset, we will need to be able to map `fipscty` to the county name.  Let us use yet another dataset to get that information.

* Source: https://wifire-data.sdsc.edu/dataset/counties-in-california/resource/248bb029-e66b-4cc1-b551-ec4b3642ea3f

```python
counties_df = pd.read_csv("./California_County_Boundaries.csv")
counties_df.head()
```

```python
counties_df.info()
```

```python
counties_df.describe()
```

```python
df = counties_df \
    .merge(gdf, how="left", left_on="County_FIPS_ID", right_index=True) \
    .copy()
df.info()
```

Let us drop columns we won't be using and convert potentially useful categorical columns.  The remaining columns wlil be left as numeric features.

```python
df = df.drop(labels=["OBJECTID", "State_FIPS_ID", "County_FIPS_ID"], axis=1)
cat_features = ["AdminRegion", "FireMAR", "LawMAR"]
for feature in cat_features:
    df[feature] = df[feature].astype("category")
df.info()
```

Now let us throw get some population numbers for each county and include that figure.

* Source: https://www.california-demographics.com/counties_by_population

```python
pop_df = pd.read_csv("./county_population.csv")
pop_df.head()
```

```python
pop_df.info()
```

```python
pop_df.describe()
```

```python
final_df = df \
    .merge(pop_df, how="left", left_on="CountyName", right_on="County") \
    .copy() \
    .drop(labels="County", axis=1)
final_df.info()
```

```python
final_df.head()
```

And let us save this as a feather file.

```python
final_df.to_feather("counties.feather")
```
